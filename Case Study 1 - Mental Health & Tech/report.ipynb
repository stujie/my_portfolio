## 2.0 Data Set

### 2.1 Source
This dataset is from a 2014 survey that measures attitudes towards mental health and frequency of mental health disorders in the tech workplace.

I located and downloaded the necessary files for this study from kaggle.com. The organization that originally uploaded this dataset to kaggle is Open Source Mental Illness (OSMI), a non-profit organization dedicated to raising awareness, educating, and providing resources to support mental wellness in the tech and open source communities.

[Link: https://www.kaggle.com/datasets/osmi/mental-health-in-tech-survey](https://www.kaggle.com/datasets/osmi/mental-health-in-tech-survey)

### 2.2 Variable Descriptions
| # | Variable | Description|
| --- | --- | --- |
| 1 | Timestamp | Date and time at which the survey response was submitted |
| 2 | Age | Respondent age |
| 3 | Gender | Respondent gender |
| 4 | Country | Respondent country of residence |
| 5 | state | If respondent lives in the United States, shows the abbreivation for their state of residence |
| 6 | self_employed | Is respondent self-employed? |
| 7 | family_history | Does respondent have a family history of mental illness? |
| 8 | treatment | Has respondent sought treatment for a mental health condition? |
| 9 | work_interfere | If respondent has mental health condition, do they feel that it interferes with their work? |
| 10 | no_employees | How many employees does the respondent's company/organization have? |
| 11 | remote_work | Does respondent work remotely (outside of office) at least 50% of the time? |
| 12 | tech_company | Is respondent's employer primarily a tech company/organization? |
| 13 | benefits | Does respondent's employer provide mental health benefits? |
| 14 | care_options | Does respondent know the options for mental health care that their employer provides? |
| 15 | wellness_program | Has respondent ever discussed mental health as part of an employee wellness program? |
| 16 | seek_help | Does respondent's employer provide resources to learn more about mental health issues and how to seek help? |
| 17 | anonymity | Is respondent's anonymity protected if they choose to take advantage of mental health or substance abuse treatment resources? |
| 18 | leave | How easy is it for respondent to take medical leave for a mental health condition? |
| 19 | mental_health_consequence | Does respondent think that discussing a mental health issue with their employer would have negative consequences? |
| 20 | phys_health_consequence | Does respondent think that discussing a physical health issue with their employer would have negative consequences? |
| 21 | coworkers | Would respondent be willing to discuss a mental health issue with their coworkers? |
| 22 | supervisor | Would respondent be willing to discuss a mental health issue with their direct supervisor(s)? |
| 23 | mental_health_interview | Would respondent bring up a mental health issue with a potential employer in an interview? |
| 24 | phys_health_interview | Would respondent bring up a physical health issue with a potential employer in an interview? |
| 25 | mental_vs_physical | Does respondent feel that their employer takes mental health as seriously as physical health? |
| 26 | obs_consequence | Has the respondent ever heard of or observed negative consequences for coworkers with mental health conditions in their workplace? |
| 27 | comments | Any additional notes or comments. |
